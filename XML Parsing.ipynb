```python
import xml.etree.ElementTree as ET
```

```python
#parsing the file
tree= ET.parse("C:\\Users\\Sanchit\\Desktop\\IGIB\\SmallData.xml")
```

```python
root= tree.getroot()
root.tag
```

```python
#printing the tags and attributes of children node
for child in root:
    print(child.tag, child.attrib)
```

```python
#printing all the tags present in file inside PubmedArticleSet 
[elem.tag for elem in root.iter()]
```

```python
#Using tostring function to call the entire xml file
print(ET.tostring(root, encoding='utf8').decode('utf8'))
```

```python
#iteration over the child node journal, output is spaces for Journal.text as it has further children tags
for Journal in root.iter('Journal'):
    print(Journal.attrib, Journal.tag, Journal.text)
```

```python
#listing the child nodes under first Journal node
for child in root.find('.//Journal'):
    print(child.tag, child.attrib)
```

```python
#the values of first child nodes under Journal
for child in root.find('.//Journal'):
    print(child.tag,":", child.text)
```

```python
#directly calling all nodes results in spaces
for child in root.findall('.//Journal'):
    print(child.tag,":", child.text)
```

```python
#calling the list of titles under Journal node 
for each in root.findall('.//Journal'):
    rating = each.find('.//Title')
    print(rating.text)
```
